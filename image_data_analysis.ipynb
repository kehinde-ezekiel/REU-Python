```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
cv2.__version__
```

```python
imag1 = cv2.imread('C:/Users/kenny/Downloads/Documents/REU-DataScienceProgram/REU-Python/forest.jpg')
imag2 = cv2.imread('C:/Users/kenny/Downloads/Documents/REU-DataScienceProgram/REU-Python/people-walking.jpg')
print('showing size image1', imag1.shape)
print('showing size image2', imag2.shape)

#number 3 states for the chanel meaning the image have 3 chanels
#the images need to have the same size in ordr to add thm
#reshape
imag1_reshaped = cv2.resize(imag1,(300,300), interpolation=cv2.INTER_AREA)
imag2_reshaped = cv2.resize(imag2,(300,300), interpolation=cv2.INTER_AREA)

print('Showing imag2 size reshaped', imag2_reshaped.shape)
```

```python
#Showing Images
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.imshow(imag1_reshaped)
plt.subplot(222)
plt.imshow(imag2_reshaped)
```

```python
#print RGB Channel
plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(imag1_reshaped[:,:,0])
plt.gca().set_title('Red channel')#insert title
plt.subplot(132)
plt.imshow(imag1_reshaped[:,:,1])
plt.gca().set_title('Green channel')#insert title
plt.subplot(133)
plt.imshow(imag1_reshaped[:,:,2])
plt.gca().set_title('blue channel')#insert title
```

```python

#add two pictures
dst = cv2.addWeighted(imag1_reshaped, 0.2, imag2_reshaped, 0.5,0.1)#why does the last one have two numb 

plt.figure(figsize=(10,10))
plt.imshow(dst)
```

```python

dst_sustraction = cv2.subtract(imag1_reshaped, imag2_reshaped)
plt.figure(figsize=(10,10))
plt.imshow(dst_sustraction)
```

```python
dst_mult = cv2.multiply(imag1_reshaped-20,imag2_reshaped-20)
plt.figure(figsize=(10,10))
plt.imshow(dst_mult)
```
